```python
dataset = pd.DataFrame(df.data)
#get entire dataset
dataset.columns = df.feature_names
print(dataset.head())
```

```python
dataset['Price'] = df.target
#create new feature 'Price', it is in the form of array.It is dependent feature.
```

```python
##Dividing the dataset into independent and dependent to solve linear regression
X = dataset.iloc[:,:-1] ##independent features by skipping last feature, skip last column
y = dataset.iloc[:,-1] ##dependednt, take the last column
```

```python
#after lasso , do train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42) #train test split with 33%
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
#we use cross validation so that every combination of train and test data is used by model, whose accuracy is better that will be combined
mse = cross_val_score(lin_reg,X_train,y_train,scoring= 'neg_mean_squared_error',cv=5) # cross validation 5 times
#we can also do train test split initially and give x train and y train here to do cross validation
#use negative mean squared error, we will get 5 different values as cv = 5
print(mse)
#here cross validation is done
```

```python
#to take avg of 5 mse
mean_mse = np.mean(mse)
print(mean_mse)
#negative value is there bcoz we have used negative mean squared error
#now it is -25. This value should go towards zero, means better performance
```

we will go with ridge and lasso regression bcoz there are lot of hyperparameters to tune in

# Ridge

```python
#Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV #for hyperparameter tuning
#define model
ridge = Ridge()
params = {'alpha' : [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,15,25,35,45,55,100]}
#apply gridsearch CV
ridge_regressor = GridSearchCV(ridge,params,scoring = 'neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)                     
```

```python
#select best parametrs after fitting
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
#after adding more parameter values
#negative mean squared error has reduced, we can not use ridge regression
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)

#linear = -37
#ridge = -29, performance is worsen
#we are not doing scaling bcoz scaling will definitely give better performance
#ridge also tries to reduce the overfitting, here performance is bad with Ridge, so try lasso regression
#after doing train test split -25.18
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
#Independent features
X = pd.DataFrame(df['data'], columns = df['feature_names'])
```

```python
X.head()
```

```python
#dependent feature is present in df.target
y = pd.DataFrame(df['target'],columns = ["Target"])
y
```

```python
#check if dataset is imbalanced, if yes we will perform upsampling
y['Target'].value_counts()
```

```python
#this is probably balanced dataset
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42) #train test split with 33%
```

```python
params = [{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
#define model
model1 = LogisticRegression(C=100, max_iter=100)
```

```python
#model with respect to gridsearchCV
model = GridSearchCV(model1,param_grid=params,scoring = 'f1',cv=5)
#we dont know id TP is important or TN so use f1
```

```python
model.fit(X_train, y_train)
```

```python
#find best param score
model.best_params_
```

```python
model.best_score_
```

```python
#now test on test data
y_pred = model.predict(X_test)
y_pred
```

```python
#after getting prediction values, apply confusion metrics
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
